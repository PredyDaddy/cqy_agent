## 1. langchain
- 主要充当的是胶水的过程，充当大模型和应用的一个工具
- 想法，行动，反馈
- 为了兼容各种大模型，langchain是没有大模型的

## 2. 能力
- 把N个不同模态的模型连接起来
- 管控用户的输入和输出，把输入更好的输入给LLMS
- 然后同时反馈给用户，然后更好的做这些东西

## 3. 主要功能
- 提示词功能
- Chain链式功能
- RAG
- Agent
- memory 
- 评估

## 4. 优缺点
- 多个模型的灵活度
- 轻量的SDk
- 多模态，图像，语言，音频
- 没办法大型工业化生产
